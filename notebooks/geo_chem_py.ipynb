```python
import csv
import random
import pandas as pd
```

```python
with open('F:\summer internship\dataset\geochem\GSNSWDataset\ASSAY_SURFSAMP_AuBasePath.csv','rt') as csv_in:
    rdr = csv.reader(csv_in)
    header = ['ID'] + next(rdr)
    rows = []
    for i, row in enumerate(rdr):
        rows.append([i+1] + row)
        
print(header)
print(len(rows))
```

```python
header_dict = {}
for idx, h in enumerate(header):
    header_dict[h] = idx
    
print(header_dict)
```

```python
def get_cell(row, column):
    idx = header_dict[column]
    return row[idx]
```

```python
def match_rows(rows, elements_subset, idx1, idx2):
    
    #print('row1,row2',idx1,idx2)
    matches = 0
    el_idx = [header_dict[el] for el in elements_subset]
    for eid in el_idx:
        if rows[idx1][eid] == rows[idx2][eid]:
            matches += 1
    #print('matches',matches)
    if matches == len(elements_subset):
        return True
    else:
        return False
            

def check_duplicates(rows, elements_subset, iteration):
    #print(elements_subset)
    matching_row_index = set()
    for idx1 , rw1 in enumerate(rows):
        for idx2, rw2 in enumerate(rows[idx1+1:], start = idx1+1):
            if match_rows(rows, elements_subset, idx1, idx2):
                matching_row_index.add((rw1[0],rw2[0]))
                
    if len(matching_row_index) != 0:
        return matching_row_index
               
    return
```

```python
num_elements = 7
min_values = 4
iterations = 5

elements_list = header[10:]

duplicates_dict = {}

for i in range(iterations):
    
    valid_rows = []

    subset = random.sample(elements_list, num_elements)
    subset.sort()
    print(subset)
    
    for rw in rows[0:5000]:
        #print("************************************************************")
        valid_count = 0
        for el in subset:
            val = get_cell(rw,el) 
            if val != '':
                if float(val) > 0:
                    valid_count += 1
                    val_float = float(val)
            else:
                continue
            #print(el, val)
        if valid_count >= min_values:
            #print(rw)
            valid_rows.append(rw)
            
    print('valid rows',len(valid_rows))
    
    
    duplicate_set = check_duplicates(valid_rows, subset, i)
    #print(duplicate_set)
    #duplicates.append(duplicate_set)
    if duplicate_set != None:
        for pair in duplicate_set:
            duplicates_dict[pair] = duplicates_dict.get(pair, 0) + 1
    else: 
        continue
#print(duplicates_dict)

sorted_duplicates = sorted(duplicates_dict, key=duplicates_dict.get, reverse=True)
for key in sorted_duplicates[:5]:
    print(key, duplicates_dict[key])
```
